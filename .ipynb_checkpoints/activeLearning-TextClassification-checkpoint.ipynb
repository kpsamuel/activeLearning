```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

import matplotlib.pyplot as plt
```

```python
## loading the dataset
import os

dataset = []

dataset_path = "dataset/bbc"
for root, files, dirs in tqdm(os.walk(dataset_path), desc="reading dataset"):
    for file in dirs:
        filepath = os.path.join(root, file).replace('\\','/')
        class_name = filepath.split('/')[-2]
            
        with open(filepath, "r") as fp:
            filedata = " ".join(fp.readlines())
        
        dataset.append({"class" : class_name, "description" : filedata})
        
dataset = pd.DataFrame(dataset)
dataset.head()
```

```python
## check the value count of each class in the dataset
dataset["class"].value_counts()
```

## simple classifier model

```python
## feature transformation
vectorizer = TfidfVectorizer(ngram_range=(2,3))
encoder = LabelEncoder()

## creating the training and test datasets
x_train, x_test, y_train, y_test = train_test_split(dataset["description"], dataset["class"], test_size=0.2)

x_train = vectorizer.fit_transform(x_train)
y_train = encoder.fit_transform(y_train)

x_test = vectorizer.transform(x_test)
y_test = encoder.transform(y_test)

print(f"feature shape {x_train.shape, y_train.shape, x_test.shape, y_test.shape}")
```

```python
## training simple KNN model

model = KNeighborsClassifier(n_neighbors=len(np.unique(y_train)))
model.fit(x_train, y_train)
pred = model.predict(x_test)

class_report = classification_report(y_test, pred)
print(class_report)
```

```python
confu_matrix = pd.DataFrame(confusion_matrix(y_test, pred))
confu_matrix.index = encoder.classes_
confu_matrix
```

## implementing the Active Learning

```python
## loading the manually labeled dataset
dataset = pd.read_csv("bbc_news.csv", encoding="iso-8859-1")
dataset.head()
```

```python
## we can check how many data points needs to be labled with the Active Learning method
dataset.isna().sum()
```

```python
## checking how many data points were manually labelled
dataset.label.value_counts()
```

```python
def transformFeatures(pooled_training_dataset, pooled_test_dataset, x_feature, y_feature):
    global encoder
    
    ## feature transformation methods
    vectorizer = TfidfVectorizer(ngram_range=(2,3), lowercase=True)
    encoder = LabelEncoder()
    
    ## fit and transform the training data points
    transformed_X_train = vectorizer.fit_transform(pooled_training_dataset[x_feature])
    transformed_y_train = encoder.fit_transform(pooled_training_dataset[y_feature])
    trained_labels = encoder.transform(pooled_training_dataset[y_feature])
    
    ## only transform the test data points
    transformed_X_test = vectorizer.transform(pooled_test_dataset[x_feature])
    
    ## label assignment startegy. used to evalulate the model training performance
    likelihood_labels = postLabelling(transformed_X_train, transformed_X_test, trained_labels)
    
    return (transformed_X_train, transformed_y_train), transformed_X_test, likelihood_labels


def postLabelling(transformed_x_train, transformed_x_test, trained_labels, method="similarity"):
    """
        # similarity : labeling mechanism based on similarity with training dataset
    """
    if method == "similarity":
        similarities = cosine_similarity(transformed_x_test, transformed_x_train)
        similar_labels = trained_labels[np.argmax(similarities, axis=1)].tolist()
        return similar_labels
```

```python
def modelEvalution(y_true, y_pred):
    model_classification_report = classification_report(y_true, y_pred, output_dict=True)
    return model_classification_report
    
    
    
def buildActiveLearningModel(model, pooled_training_dataset, pooled_test_dataset, x_feature, y_feature):
    
    ## transforming the training and prediction data points
    (x_train, y_train), x_test, likelihood_labels = transformFeatures(pooled_training_dataset, pooled_test_dataset, x_feature, y_feature)
    
    ## model training, predicting with probability. NOTE : model should support probability score
    model.fit(x_train, y_train)
    pred_probs = model.predict_proba(x_test)
    pred_label = np.argmax(pred_probs, axis=1)
    
    ## mapping the predictions to test data points
    pooled_test_dataset["likelihood_label"] = likelihood_labels
    pooled_test_dataset.loc[pooled_test_dataset.index, "label"] = pred_label
    pooled_test_dataset.loc[pooled_test_dataset.index, "label_score"] = np.max(pred_probs, axis=1)
    
    ## mapping each class scores to test data points
    for label_id in range(pred_probs.shape[1]):
        label_name = f"label_{label_id}"
        label_score = pred_probs[:,label_id]
        pooled_test_dataset.loc[pooled_test_dataset.index, label_name] = label_score
            
    ## checking if model predictions are confusions or not.
    ## model confusion is seen when model gives same probability scores to more than one class
    pooled_test_dataset["prediction_confusion"] = False
    prediction_confusions = np.array([np.count_nonzero(pred_probs[i] == pred_probs[i].max()) for i in range(pred_probs.shape[0])])
    prediction_confusions_index = list(np.where(prediction_confusions > 1)[0])
    pooled_test_dataset.loc[prediction_confusions_index, "prediction_confusion"] = True
    
    ## checking the model performance
    evalution_results = modelEvalution(y_true=likelihood_labels, y_pred=pred_label)
    
    return pooled_test_dataset, evalution_results
```

```python
## dataset preparation
training_dataset = dataset[~dataset["label"].isna()]           ## pre-labeled dataset
non_labeled_dataset = dataset[dataset["label"].isna()]         ## need to label dataset

training_dataset.reset_index(inplace=True, drop=True)
non_labeled_dataset.reset_index(inplace=True, drop=True)
print(f"initial label data : {training_dataset.shape} non label data : {non_labeled_dataset.shape}")


## start active learning
num_experiments = 10
all_experiment_logs = []


for experiment_id in tqdm(range(num_experiments), desc="experimentation started"):
    
    ## tracking the pre-experiment logs
    experiment_logs = {
        "experiment_id" : experiment_id,
        "training_datset" : training_dataset.shape[0],
        "non_labeled_dataset" : non_labeled_dataset.shape[0]
    }
    
    ## model training and predicting
    results, evalution_results = buildActiveLearningModel(model = KNeighborsClassifier(), 
                                       pooled_training_dataset = training_dataset, 
                                       pooled_test_dataset = non_labeled_dataset,
                                       x_feature = "description",
                                       y_feature = "label")

    # 1. labeling method - least confident : label only the most confident and leave the rest to be learnt correctly

    least_confidence_threshold = 0.4
    needs_to_label = results[results["label_score"] <= least_confidence_threshold]
    non_labeled_dataset = needs_to_label
    non_labeled_dataset.reset_index(inplace=True, drop=True)
    
    
    ## adding the well learnt data point for training
    model_learnt = results[results["label_score"] > least_confidence_threshold]
    model_learnt["label"] = encoder.inverse_transform(model_learnt["label"])
    training_dataset = pd.concat([training_dataset, model_learnt[training_dataset.columns]], axis=0)
    training_dataset.reset_index(inplace=True, drop=True)
    
    ## tracking the post-experiment logs
    experiment_logs.update({
            "least_confidence_threshold" : least_confidence_threshold,
            "learnt_dataset" : model_learnt.shape[0]
    })
    
    experiment_logs.update(evalution_results)
    all_experiment_logs.append(experiment_logs)
    
    
print(f"after label data : {training_dataset.shape} | after non label data : {non_labeled_dataset.shape}")
```

## check the experiment logs

```python
metric_name = "precision"

fig = plt.figure(figsize=(10,5))
for exper_logs in all_experiment_logs:
    experiment_id = exper_logs["experiment_id"]
    exper_logs = pd.DataFrame(exper_logs)
    plt.plot(exper_logs[["0","1","2","3","4"]].loc[metric_name], label=f"experiment-{experiment_id}")
    
plt.title(f"{metric_name} of each classes in all the experiments")
plt.ylabel(f"{metric_name}-score")
plt.xlabel("classes")
plt.legend(loc="upper left")
plt.show()
```
